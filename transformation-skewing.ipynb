```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('big.jpg')
```

```python
rows, cols, ch = img.shape    
pts1 = np.float32(
    [[cols*.25, rows*.75],
     [cols*.75, rows*.75],
     [cols*.75, 0],
     [cols,     0]]
)
pts2 = np.float32(
    [[cols*0.1, rows],
     [cols,     rows],
     [0,        0],
     [cols,     0]]
)    
print(pts1, pts2)
```

```python
M = cv2.getPerspectiveTransform(pts1,pts2)
skew_image = cv2.warpPerspective(img, M, (cols, rows))
plt.imshow(skew_image)
```
