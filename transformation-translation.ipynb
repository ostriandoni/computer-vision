```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('big.jpg')
```

```python
# shift the image 100 pixels to the right and 250 pixels down
shift_coordinate = np.float32([[1, 0, 100], [0, 1, 250]])
print(shift_coordinate)
```

```python
shifted_image = cv2.warpAffine(image, shift_coordinate, (image.shape[1], image.shape[0]))
plt.imshow(shifted_image)
```
